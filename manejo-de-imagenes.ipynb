```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from skimage import io
```

```python
im = io.imread('/kaggle/input/cnn-data-sources/Mara.jpeg')
```

```python
im.shape # Una imagen de 800 x 600 y 3 capas de colores 
```

```python
plt.imshow(im)
```

```python
roja = im[:,:,0]# Capa cero que es la capa   roja
verde = im[:,:,1]
azul = im[:,:,2]
```

```python
plt.imshow(azul,cmap='gray')
```

```python
# como las imagenes son matrices podemos manipularlas con numpy
plt.imshow(roja.T)
```

```python
#  creamos variable auxiliar de zeros con la misma dimension que la imagen
aux_dim = np.zeros([800,600])
```

```python
r = np.dstack((roja,aux_dim,aux_dim)).astype(np.uint8) # Hcemos que la matriz solo tenga la informacion en rojo (1er capa) y la demas sera cero
g = np.dstack((aux_dim,verde,aux_dim)).astype(np.uint8) # Hcemos que la matriz solo tenga la informacion en verde (2da capa) y la demas sera cero
b = np.dstack((aux_dim,aux_dim,azul)).astype(np.uint8) # Hcemos que la matriz solo tenga la informacion en azul (3er capa) y la demas sera cero
```

```python
#visualizamos el resultado
plt.imshow(g)
```
