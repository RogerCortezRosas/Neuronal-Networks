```python
import numpy as np
import matplotlib.pyplot as plt 
import scipy.ndimage as nd 
from skimage import  io, color 
```

## Carga de imagen

```python
img = io.imread('/kaggle/input/cnn-data-sources/yo.png')
```

```python
# Pasamos la imagen a escala de grises para mejor procesamiento
img_gray = color.rgb2gray(img)
```

```python
plt.imshow(img_gray)
```

# Creando kernels

```python
#kernel blur -> El núcleo de desenfoque resta importancia a las diferencias en los valores de los píxeles adyacentes.

blur = np.array([[0.0625,0.125,0.0625],
                   [0.125,0.25,0.125],
                   [0.0625,0.125,0.0625]])

# kernel bottom sobel -> Los núcleos Sobel se utilizan para mostrar solo las diferencias en los valores de píxeles adyacentes en una dirección particular.

sobel = np.array([[-1,-2,-1],
                   [0,0,0],
                   [1,2,1]])

# Kernel custom -> Se queda exactamente igual la imagen

custom = np.array([[0,0,0],
                   [0,1,0],
                   [0,0,0]])

# Kernel vertical - > Solo muestra las fromas o bordes verticales

vertical = np.array([[-1,0,1],
                   [-1,0,1],
                   [-1,0,1]])

# Kernel outline o edge -> Se utiliza un núcleo de contorno (también llamado núcleo de "borde") para resaltar grandes diferencias en los valores de los píxeles. Un píxel que esté junto a píxeles vecinos con una intensidad similar aparecerá en negro en la nueva imagen, mientras que uno que esté junto a píxeles vecinos que difieran mucho aparecerá en blanco.

edge = np.array([[-1,-1,-1],
                   [-1,8,-1],
                   [-1,-1,-1]])
```

```python
# Convoluciones

blur_conv = nd.convolve(img_gray, blur)
sobel_conv = nd.convolve(img_gray, sobel)
custom_conv = nd.convolve(img_gray, custom)
vertical_conv = nd.convolve(img_gray, vertical)
edge_conv = nd.convolve(img_gray, edge)
```

```python
fig, axes = plt.subplots(2,3, figsize = (15,10))
axes[0,0].imshow(img_gray, cmap=plt.cm.gray)
axes[0,0].set_title('Original')
axes[0,0].axis('off')

axes[0,1].imshow(blur_conv, cmap=plt.cm.gray)
axes[0,1].set_title('blur')
axes[0,1].axis('off')

axes[0,2].imshow(sobel_conv, cmap=plt.cm.gray)
axes[0,2].set_title('sobel')
axes[0,2].axis('off')

axes[1,0].imshow(custom_conv, cmap=plt.cm.gray)
axes[1,0].set_title('custom')
axes[1,0].axis('off')

axes[1,1].imshow(vertical_conv, cmap=plt.cm.gray)
axes[1,1].set_title('vertical')
axes[1,1].axis('off')

axes[1,2].imshow(edge_conv, cmap=plt.cm.gray)
axes[1,2].set_title('edge')
axes[1,2].axis('off')



plt.show()
```
